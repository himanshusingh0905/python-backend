### short circuiting in python
#### 1. Short Circuit with `or`

`or` stops when it finds the **first truthy value**.

**Rule**
```py
A or B
```


- If **A is truthy → return A (B is NOT evaluated)**
- If **A is falsy → evaluate and return B**

**Example**

```python
a = 10
b = 5

result = a or b
print(result)  # output = 10, Because a is already truthy, Python never checks b.
```
---

### 2. Short Circuit with `and`

`and` stops when it finds the **first falsy value**.

**Rule**
```py
A or B
```


- If **A is falsy → return A (B not evaluated)**
- If **A is truthy → evaluate and return B**

**Example**

```python
a = 0
b = 5

result = a and b
print(result) # Output : 0, b is never evaluated.
```


````python
den = 0
nume = 8

safe_division = den and nume / den

print(safe_division)
````

### Enum:
- auto()
- strEnum class
- Enum class

````python
from enum import Enum

class OrderStatus(Enum):
    PENDING   = "pending"
    CONFIRMED = "confirmed"
    SHIPPED   = "shipped"
    CANCELLED = "cancelled"


# Accessing a member
status = OrderStatus.PENDING
print(status)

# The member has two parts:
print(status.name)    # "PENDING"  ← the attribute name
print(status.value)   # "pending"  ← the assigned value

# Comparison — always compare member to member, not to raw string
print(status == OrderStatus.PENDING)    # ✅ True, Both sides are Enum members of the same Enum class
print(status == "pending" )             # ❌ False — status is an Enum member, not the string "pending".

````

### auto() — when the value doesn't matter
- Sometimes you have a set of states where the actual underlying value is ***irrelevant*** — you just need distinct members. 
- auto() generates values for you automatically so you don't invent meaningless numbers.

Means you have no use of value in your logic, you will only use name.


````python
from enum import Enum, auto

class Direction(Enum):
    NORTH = auto()   # gets value 1
    SOUTH = auto()   # gets value 2
    EAST  = auto()   # gets value 3
    WEST  = auto()   # gets value 4


# 4 enum members in above class
````

### StrEnum — the production workhorse
- This is the one you'll use most in real backend/API code. 
- `StrEnum` makes the *enum member* behave like a string — it compares equal to its value, works in string contexts, and serialises cleanly to JSON.

````python
from enum import StrEnum, auto

class OrderStatus(StrEnum):
    PENDING   = auto()   # value becomes "pending" (lowercased name)
    CONFIRMED = auto()   # value becomes "confirmed"
    SHIPPED   = auto()   # value becomes "shipped"
    CANCELLED = auto()   # value becomes "cancelled"


status = OrderStatus.PENDING

# StrEnum members ARE strings — they pass string checks
print(status)               # "pending"
print(status.value)         # "pending"
print(status == "pending" )        # ✅ True  ← key difference from plain Enum
print(isinstance(status, str) )    # ✅ True


````

````python
# Serialises to JSON without any extra work

import json
data = {"status": OrderStatus.SHIPPED}
print(json.dumps(data))     # '{"status": "shipped"}' ✅

# Works in f-strings, string formatting, everywhere a str works
print(f"Order is {OrderStatus.CONFIRMED}")  # "Order is confirmed"
````

## core difference b/w Enum and StrEnum:
- `Enum` member is NOT its value. `StrEnum` member IS its value.

```py
from enum import Enum, StrEnum, auto

class StatusEnum(Enum):
    PENDING = "pending"

class StatusStrEnum(StrEnum):
    PENDING = "pending"

# --- Comparison ---
StatusEnum.PENDING == "pending"      # ❌ False
StatusStrEnum.PENDING == "pending"   # ✅ True

# --- isinstance check ---
isinstance(StatusEnum.PENDING, str)      # ❌ False
isinstance(StatusStrEnum.PENDING, str)   # ✅ True
```
